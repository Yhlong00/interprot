```python
import pandas as pd 
import matplotlib.pyplot as plt

info = pd.read_csv('sc.csv')
fig, axs = plt.subplots(2, 4, figsize=(16, 8))

# Distribution plots with median line
for i, column in enumerate(['seqlen', 'seqid', 'scrmsd', 'sctm']):
    counts, bins, patches = axs[0, i].hist(info[column], bins=10, color=['blue', 'green', 'red', 'purple'][i], alpha=0.7, edgecolor='black')
    
    axs[0, i].set_title(f'Distribution of {column}')
    
    # Set xticks to the start and end of each bin
    axs[0, i].set_xticks(bins[::2])
    
    # Adding a median line
    median_val = info[column].median()
    axs[0, i].axvline(median_val, color='orange', linestyle='dashed', linewidth=2)
    axs[0, i].text(median_val, axs[0, i].get_ylim()[1] * 0.9, f'Median: {median_val:.2f}', color='orange')

# Scatter plots (Second row)
axs[1, 0].scatter(info['seqlen'], info['seqid'], color='blue', alpha=0.7)
axs[1, 0].set_xlabel('seqlen')
axs[1, 0].set_ylabel('seqid')
axs[1, 0].set_title('seqlen vs seqid')

axs[1, 1].scatter(info['seqlen'], info['scrmsd'], color='green', alpha=0.7)
axs[1, 1].set_xlabel('seqlen')
axs[1, 1].set_ylabel('scrmsd')
axs[1, 1].set_title('seqlen vs scrmsd')

axs[1, 2].scatter(info['seqlen'], info['sctm'], color='red', alpha=0.7)
axs[1, 2].set_xlabel('seqlen')
axs[1, 2].set_ylabel('sctm')
axs[1, 2].set_title('seqlen vs sctm')

# Hide the last empty subplot
fig.delaxes(axs[1, 3])

# Adjust layout
plt.tight_layout()
plt.show()
```
